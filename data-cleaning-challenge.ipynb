```python
# libraries
import pandas as pd

# Present Working Directory | Get Current Working Directory
import os
pwd = os.getcwd()
pwd
```

```python
# import csv into jupyter notebooks
data_import = pd.read_csv(pwd + '\\data_cleaning_challenge.csv')
data_import
```

```python
# change header names to Row Type, Iter Number, Power1, etc. 
# this was a manual manipulatin to the CSV
```

```python
# drop empty columns - using 'columns = []', you can drop multiple columns with a list
# create a new variable for a new copy before dropping, so that the existing data set can remain as the source
drop_extra = data_import.drop(columns = ['Unnamed: 9', 'Unnamed: 10'])
drop_extra
```

```python
# drop NaN values - rows drop from, 76377 to 58397
drop_nans = drop_extra[drop_extra['Row Type'].notna()]
drop_nans
```

```python
# create a for loop that searches for 'first name' and then it copies all data until it reaches 'first name' again
# create a list
column_values = []
counter = 0

for i in drop_nans['Row Type']:
    if 'first name' in i:
        # print(i) used for error handling
        counter += 1
    column_values.append(counter)
        
```

```python
# we now have a list for our column values
# we want to make this a new column in our dataset
iter_cols = drop_nans
iter_cols['Iteration'] = column_values
iter_cols
```

```python
# view current variables
%who_ls
```

```python
# set rows displayed
pd.set_option('display.max_rows', 100)
iter_cols.head(100)
```

```python
# reset rows displayed
pd.reset_option('all')
```

```python
# remove 'index 2' row - unneeded row
drop_extra_columns_name_rows = iter_cols[iter_cols['Row Type'] != 'Row Type']
drop_extra_columns_name_rows
```

```python
# seperate columns from data frame, creating a 'name data frame'
name_dataframe = drop_extra_columns_name_rows[drop_extra_columns_name_rows['Row Type'].str.contains('first name')]

# using inplace, the .drop changes will be committed to the dataframe
name_dataframe.drop(columns = ['Speed1', 'Speed2', 'Electricity', 'Effort', 'Weight', 'Torque'], inplace = True)

# create dictionary to change the header names
name_dataframe.rename(columns = {'Row Type': 'First Name', 'Iter Number': 'Last Name', 'Power1': 'Date'}, inplace = True)

# remove 'first name', 'last name', from the values in the rows
# string slicing - take the first few indices of the overall value
name_dataframe['First Name'] = name_dataframe['First Name'].str[12:]  # start at 12 index and take the values after
name_dataframe['Last Name'] = name_dataframe['Last Name'].str[11:] # start at 11 index and take the values after
name_dataframe['Date'] = name_dataframe['Date'].str[6:] # start at 6 index and take the values after

name_dataframe
```

```python
# create a new dataframe without all of this first name information - a 'no name dataframe'
# want the complete opposite of the name_dataframe - you can use an inverse opertaor'~' 
# this will pull exact opposite rows from name_dataframe logic
no_name_dataframe = drop_extra_columns_name_rows[~drop_extra_columns_name_rows['Row Type'].str.contains('first name')]
no_name_dataframe
```

```python
# merge these two dataframes together
merged_dataframe = pd.merge(left = name_dataframe, right = no_name_dataframe, how = 'inner', on = 'Iteration')
merged_dataframe
```

```python
# drop 'Iteration' column
drop_iter_col_merged_dataframe = merged_dataframe.drop(columns = ['Iteration'])
drop_iter_col_merged_dataframe
```

```python
# export to csv and remove index
drop_iter_col_merged_dataframe.to_csv(pwd + '\\final-data_cleaning_challenge.csv', index = False)
```
